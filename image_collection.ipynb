```python
import cv2
import os
import time

def collect_image(person_name: str) -> None:
    path = 'Dataset'
    total_images = 100

    if not os.path.isdir(path):
        os.makedirs(path)

    name_dir = os.path.join(path, person_name)
    if not os.path.isdir(name_dir):
        os.makedirs(name_dir)

    camera = cv2.VideoCapture(0)
    if not camera.isOpened():
        print("The web camera is not accessible.")
        return

    print(f"Collecting {total_images} images for: {person_name}")
    count = 0

    while count < total_images:
        ret, frame = camera.read()
        if not ret:
            print("Failed to grab frame.")
            break

        # cv2.imshow("Image Collection - Press 'q' to quit", frame)

        img_path = os.path.join(name_dir, f'image_{count+1}.jpg')
        cv2.imwrite(img_path, frame)
        count += 1
        print(f"Saved image {count}/{total_images}")

        time.sleep(0.1)

        # if cv2.waitKey(1) & 0xFF == ord('q'):
        #     print("Exiting early...")
        #     break

    print(f"Completed image collection for {person_name}.")
    camera.release()
    # cv2.destroyAllWindows()
```

```python
total_person = int(input("Enter total people for attendence"))
```

```python
for person in range(1, total_person+1):
    name = input("Enter the name of the Person : ")
    collect_image(name)
```
