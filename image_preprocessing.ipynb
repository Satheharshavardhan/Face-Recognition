```python
import os
import cv2
import numpy as np
from rembg import remove
```

```python
def process_images(input_dir='Dataset', output_dir='Processed', size=(120, 120)):
    if not os.path.isdir(output_dir):
        os.makedirs(output_dir)

    # Loop over each person
    for person_name in os.listdir(input_dir):
        person_input_path = os.path.join(input_dir, person_name)
        person_output_path = os.path.join(output_dir, person_name)

        if not os.path.isdir(person_input_path):
            continue

        os.makedirs(person_output_path, exist_ok=True)

        for img_file in os.listdir(person_input_path):
            img_path = os.path.join(person_input_path, img_file)
            output_path = os.path.join(person_output_path, img_file)

            try:
                # Read image
                image = cv2.imread(img_path)

                # Convert BGR to RGB for rembg
                image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                # Remove background
                removed = remove(image_rgb)

                # Convert back to BGR
                removed_bgr = cv2.cvtColor(np.array(removed), cv2.COLOR_RGB2BGR)

                # Resize to (120, 120)
                resized = cv2.resize(removed_bgr, size)

                # Save processed image
                cv2.imwrite(output_path, resized)
                print(f"Processed {img_file} for {person_name}")
            except Exception as e:
                print(f"Failed to process {img_file}: {e}")
```

```python
process_images()
```
